# Group and Summarize Sales

```python
#Total Sales by Region
region_sales = df.groupby('Region')['Sales'].sum()
print(region_sales)
```

```python
#Total Sales by Category
category_sales = df.groupby('Category')['Sales'].sum()
print(category_sales)

```

```python
#Most Sold Products by Quantity
top_products = df.groupby('Product')['Quantity'].sum().sort_values(ascending=False)
print(top_products)
```

# Visualize Insights

```python
#Bar Chart: Sales by Category
plt.figure(figsize=(8, 6))
category_sales.plot(kind='bar')
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.show()
```

```python
#Line Chart: Daily Sales Trend
df['Date'] = pd.to_datetime(df['Date'])
daily_sales = df.groupby('Date')['Sales'].sum()
daily_sales.plot(kind='line', marker='o', title='Daily Sales Trend', figsize=(10,5))
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

#Filter the Data


```python
#Orders with Sales Above ₹30,000
high_sales = df[df['Sales'] > 30000]
print(high_sales)
```

```python
#Clothing Items Sold in the North Region
north_clothing = df[(df['Region'] == 'North') & (df['Category'] == 'Clothing')]
print(north_clothing)
```
